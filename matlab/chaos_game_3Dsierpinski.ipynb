# Chaos Game – Sierpiński Triangle

This notebook demonstrates the *chaos game*, a simple random process that generates the Sierpiński triangle fractal.

**Idea:**
1. Start with an equilateral triangle.
2. Pick a starting point inside the triangle.
3. Repeatedly choose a random vertex and move the point halfway towards it.
4. Plot all the points – a fractal pattern emerges.


## 1. Imports

We only need NumPy for numerical operations and Matplotlib for plotting.

````python
import numpy as np
import matplotlib.pyplot as plt
````

## 2. Define the chaos game function

The function below:
- Defines an equilateral triangle.
- Starts from the centroid of the triangle.
- Iteratively applies the chaos game rule.
- Plots the resulting points.


````python
def chaos_game(num_points: int = 20000):
    """Generate and plot the Sierpiński triangle using the chaos game.

    Parameters
    ----------
    num_points : int, optional
        Number of points to generate, by default 20000.
    """

    # 1. Define vertices of an equilateral triangle
    A = np.array([0.0, 0.0])
    B = np.array([1.0, 0.0])
    C = np.array([0.5, np.sqrt(3) / 2])  # height of equilateral triangle

    vertices = np.array([A, B, C])

    # 2. Starting point: centroid of the triangle
    current_point = vertices.mean(axis=0)

    # Pre-allocate array of points for efficiency
    points = np.zeros((num_points, 2))

    # 3. Chaos game iterations
    for i in range(num_points):
        # Randomly pick one of the three vertices (index 0, 1, or 2)
        idx = np.random.randint(0, 3)
        chosen_vertex = vertices[idx]

        # Move halfway towards the chosen vertex
        current_point = (current_point + chosen_vertex) / 2.0

        # Store the new point
        points[i] = current_point

    # 4. Plot the result
    plt.figure(figsize=(5, 5))
    plt.scatter(points[:, 0], points[:, 1], s=0.2)
    plt.axis("equal")
    plt.axis("off")
    plt.title(f"Chaos Game – Sierpiński Triangle ({num_points} points)")
    plt.show()
````

## 3. Run the chaos game

Now we call the function with a chosen number of points. Try increasing this value
to see the fractal become clearer (e.g. 5000, 20000, 50000).


````python
# Example: generate 30,000 points
chaos_game(30000)
````

## 4. Experiments
