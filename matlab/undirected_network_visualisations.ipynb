# Undirected Network Visualisations (N = 100)

This notebook visualises **undirected networks** (nodes + edges) for five canonical
degree distributions in network science.

Focus: **visual intuition** â€” how hubs and structure emerge.


```python
%pip install networkx
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

N = 100
```

## 1) Regular (Uniform) Network

```python
k = 6
G = nx.random_regular_graph(k, N)
pos = nx.circular_layout(G)
nx.draw(G, pos, node_size=60, with_labels=False)
plt.title('Regular Network')
plt.show()
```

## 2) Poisson (Random / ErdÅ‘sâ€“RÃ©nyi) Network

```python
k = 6
G = nx.erdos_renyi_graph(N, k/(N-1))
pos = nx.spring_layout(G, seed=1)
nx.draw(G, pos, node_size=60)
plt.title('Poisson Network')
plt.show()
```

## 3) Exponential Network

```python
k = 6
deg = np.maximum(1, np.random.exponential(k, N).astype(int))
G = nx.configuration_model(deg, create_using=nx.Graph)
G.remove_edges_from(nx.selfloop_edges(G))
pos = nx.spring_layout(G, seed=2)
nx.draw(G, pos, node_size=60)
plt.title('Exponential Network')
plt.show()
```

## 4) Scale-Free (Power-Law) Network

```python
G = nx.barabasi_albert_graph(N, m=3)
pos = nx.spring_layout(G, seed=3)
nx.draw(G, pos, node_size=60)
plt.title('Scale-Free Network')
plt.show()
```

## 5) Log-Normal Network

```python
deg = np.random.lognormal(mean=1, sigma=0.8, size=N).astype(int)
G = nx.configuration_model(deg, create_using=nx.Graph)
G.remove_edges_from(nx.selfloop_edges(G))
pos = nx.spring_layout(G, seed=4)
nx.draw(G, pos, node_size=60)
plt.title('Log-Normal Network')
plt.show()
```
