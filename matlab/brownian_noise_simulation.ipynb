# Brownian Noise Simulation (Beginner MATLAB → Python Translation)
This notebook recreates a simple Brownian noise simulation. It mirrors the first MATLAB lesson many students encounter — generating random noise, cumulatively summing it, and plotting the result.


````python
import numpy as np
import matplotlib.pyplot as plt
````

## Generate Brownian Noise
Equivalent to MATLAB:
```matlab
R = randn(1000,50)
K = cumsum(R)
plot(K)
```

````python
R = np.random.randn(1000, 50)
K = np.cumsum(R, axis=0)

plt.figure(figsize=(10,5))
plt.plot(K)
plt.title('Brownian Noise Simulation')
plt.xlabel('Step')
plt.ylabel('Value')
plt.show()
````

## Reflection
- This exercise demonstrates how random noise behaves when accumulated.
- It helped me understand the difference between *instantaneous randomness* and *persistent drift*.
