# Degree Distribution Networks (Linear Axes)

This notebook demonstrates **five canonical degree distributions** in network science,
with **minimal Python code** and **linear-axis histograms**, suitable for teaching,
portfolio work, and GitHub.

Distributions covered:
1. Regular (Uniform)
2. Poisson (Random / Erdős–Rényi)
3. Exponential
4. Scale-Free (Power-Law)
5. Log-Normal


```python
%pip install networkx
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

N = 500
k = 6
```

## 1) Regular (Uniform) Degree Distribution

```python
G = nx.random_regular_graph(k, N)
plt.hist([d for _,d in G.degree()])
plt.title('Regular Degree Distribution')
plt.show()
```

## 2) Poisson (Random / Erdős–Rényi)

```python
G = nx.erdos_renyi_graph(N, k/(N-1))
plt.hist([d for _,d in G.degree()])
plt.title('Poisson Degree Distribution')
plt.show()
```

## 3) Exponential Degree Distribution

```python
deg = np.maximum(1, np.random.exponential(k, N).astype(int))
G = nx.configuration_model(deg, create_using=nx.Graph)
G.remove_edges_from(nx.selfloop_edges(G))
plt.hist([d for _,d in G.degree()])
plt.title('Exponential Degree Distribution')
plt.show()
```

## 4) Scale-Free (Power-Law)

```python
G = nx.barabasi_albert_graph(N, m=3)
plt.hist([d for _,d in G.degree()])
plt.title('Scale-Free Degree Distribution')
plt.show()
```

## 5) Log-Normal Degree Distribution

```python
deg = np.random.lognormal(mean=1, sigma=0.8, size=N).astype(int)
G = nx.configuration_model(deg, create_using=nx.Graph)
G.remove_edges_from(nx.selfloop_edges(G))
plt.hist([d for _,d in G.degree()])
plt.title('Log-Normal Degree Distribution')
plt.show()
```
