# Example Notebook

This notebook demonstrates a simple workflow:
1. Import a CSV dataset.
2. Display a summary.
3. Plot the data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('../data/sample.csv')

# Display the DataFrame
display(data)

# Plot the values
plt.plot(data['id'], data['value'], marker='o')
plt.title('Sample Data Plot')
plt.xlabel('ID')
plt.ylabel('Value')
plt.show()
```
