```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("../data/sample.csv")

# Prepare features and target
X = df[["x"]]
y = df["y"]

# Create and fit model
model = LinearRegression()
model.fit(X, y)

# Predict
df["predicted_y"] = model.predict(X)

# Plot results
plt.scatter(df["x"], df["y"], label="Actual")
plt.plot(df["x"], df["predicted_y"], color="red", label="Predicted")
plt.legend()
plt.title("Linear Regression Example")
plt.show()

print(f"Model Coefficient: {model.coef_[0]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")
# Save results
df.to_csv("../data/sample_with_predictions.csv", index=False)  
```
